```python
# Probando plotnine
from plotnine import * # este hau que incluirlo cuando trabajemos con jupyter
from plotnine.data import mtcars
```

```python
(
    ggplot(mtcars, aes("wt", "mpg"))
    + geom_point()
)
```

```python
(
    ggplot(mtcars, aes("wt", "mpg", color="factor(gear)"))
    + geom_point()
)
```

```python
(
    ggplot(mtcars, aes("wt", "mpg", color="factor(gear)"))
    + geom_point()
    + stat_smooth(method="lm")
)
```

```python
(
    ggplot(mtcars, aes("wt", "mpg", color="factor(gear)"))
    + geom_point()
    + stat_smooth(method="lm")
    + facet_wrap("gear")
)
```

```python
(
    ggplot(mtcars, aes("wt", "mpg", color="factor(gear)"))
    + geom_point()
    + stat_smooth(method="lm")
    + facet_wrap("gear")
    + theme_xkcd()
)
```

```python
(
    ggplot(mtcars, aes("wt", "mpg", color="factor(gear)"))
    + geom_point()
    + stat_smooth(method="lm")
    + facet_wrap("gear")
    + theme_tufte()
)
```
